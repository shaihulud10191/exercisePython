To do:
Create a csv file with columns InvoiceDt, stockCode, Description, Total sales
Perform imputation (if a product was not sold on a date then there must a zero in total sales)

```python
import pandas as pd 
origin = r"D:\Python\task\simplerName.xlsx" #assigning the raw path to a var
```

```python
df = pd.read_excel(origin) #reading the file
```

```python
df
```

```python
dfNew = df[['InvoiceDate', 'StockCode', 'Description', 'Quantity', 'Price']] #creating a new dataframe that is a subset of the original one
dfNew.insert(loc = 5, column = 'TotalSales', value = 0) #adding a new column
```

```python
dfNew
```

```python
if df['Quantity'].isna().any(): 
    print("NaN under quantity")
if (df['Quantity'] == 0).any():
    print("Zeroes under quantity") 
#output looks clear, assume quantity column is clean
if df['Price'].isna().any():
    print("NaN under price")
if (df['Price'] == 0).any():
    print("Zeroes under price df") 
#even though the problem statement says so, there aren't any days on which a product WASN'T sold
if dfNew['Quantity'].isna().any():
    print("NaN under quantity")
if (dfNew['Quantity'] == 0).any():
    print("Zeroes under quantity") 
if dfNew['Price'].isna().any():
    print("NaN under price")
if (dfNew['Price'] == 0).any():
    print("Zeroes under price") 
#same conclusions follow for the new dataframe, obviously
```

```python
dfNew['TotalSales'] = dfNew['Quantity'] * dfNew['Price'] #sales equals quantity of a product sold times price
```
